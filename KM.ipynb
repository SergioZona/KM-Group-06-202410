```python
from graphviz import Digraph

def create_decision_tree():
    dot = Digraph(comment='Decision Tree')

    # Set node attributes
    dot.attr('node', shape='box', fontsize='8', style='filled', fillcolor='lightblue', fontname='Arial')

    # Add nodes with spaced text
    dot.node('A', 'Tipo de Aplicación\n\n', fontsize='8')
    dot.node('B', 'Naturaleza del Problema\n\n', fontsize='8')
    dot.node('C', 'Etapa del Ciclo de Vida del Desarrollo del Software\n\n', fontsize='8')
    dot.node('D', 'Complejidad del Sistema o Módulo Bajo Prueba\n\n', fontsize='8')

    # Add edges
    dot.edge('A', 'B', 'Aplicación Web-Front\n\n', fontsize='8')
    dot.edge('A', 'B', 'Aplicación Móvil\n\n', fontsize='8')
    dot.edge('A', 'B', 'Software/Backend de Escritorio\n\n', fontsize='8')

    dot.edge('B', 'C', 'Fallo de Funcionalidad\n\n', fontsize='8')
    dot.edge('B', 'C', 'Rendimiento Lento\n\n', fontsize='8')
    dot.edge('B', 'C', 'Problemas de Seguridad\n\n', fontsize='8')
    dot.edge('B', 'C', 'Consultas o dudas de implementación\n\n', fontsize='8')

    dot.edge('C', 'D', 'Desarrollo Inicial\n\n', fontsize='8')
    dot.edge('C', 'D', 'Pruebas\n\n', fontsize='8')
    dot.edge('C', 'D', 'Implementación\n\n', fontsize='8')
    dot.edge('C', 'D', 'DevOps\n\n', fontsize='8')

    dot.edge('D', 'END', 'Baja\n\n', fontsize='8')
    dot.edge('D', 'END', 'Media\n\n', fontsize='8')
    dot.edge('D', 'END', 'Alta\n\n', fontsize='8')

    # Render the decision tree
    return dot

# Render decision tree directly in the notebook
decision_tree = create_decision_tree()
decision_tree.render('decision_tree', format='png', cleanup=True)
decision_tree
```
